```python
import pandas as pd
```

```python
df_natalite =  pd.read_excel('taux_natalite.xlsx')
departement_code= [16,17,19,23,24,33,40,47,64,79,86,87]
df_natalite
```

```python
df_natalite.loc[len(df_natalite.index)] = [2024,9.5] 
df_natalite=df_natalite.sort_values("Année")
```

```python
years = df_natalite["Année"]
natalite = df_natalite["Taux de natalité"]
df_nata_clean =  pd.DataFrame(columns=['Année','Code du département',"Taux de natalité"])
```

```python
for i in range(len(years)):
    for y in range (len(departement_code)):
        row = {"Année":years[i],"Code du département":departement_code[y],"Taux de natalité":natalite[i]}
        df_nata_clean.loc[len(df_nata_clean)] = row
df_nata_clean
```

```python
df_nata_clean.to_csv("natalite.csv")
```
