```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
df = pd.read_excel("taux-pauvrete.xlsx",skiprows=3)
df= df.iloc[0:4,:]
df
```

```python
#df=df.melt(['Seuil de pauvreté'], var_name='Année',value_name="t")
df = df.set_index('Seuil de pauvreté').transpose()
df['Année'] = df.index
df=df.reset_index(drop=True)
```

```python
df=df.rename(columns={"Seuil à 40 %":"taux_pauvrete_seuil_40%","Seuil à 50 %":"taux_pauvrete_seuil_50%","Seuil à 60 %":"taux_pauvrete_seuil_60%","Seuil à 70 %":"taux_pauvrete_seuil_70%"})
df
```

```python
df.columns
```

```python
departement_code= ['16','17','19','23','24','33','40','47','64','79','86','87']
df_passing_dep = []
for dep in departement_code :
    df_dep = df.copy()
    df_dep["Code du département"] = dep
    df_passing_dep.append(df_dep)
len(df_passing_dep)
```

```python
df_passing_dep[6]
```

```python
df_concat =  pd.concat(df_passing_dep)
df_concat
```

```python
df_concat.to_csv("taux-pauvrete.csv")
```
