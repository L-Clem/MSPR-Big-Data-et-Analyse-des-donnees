```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
df = pd.read_excel("rapport-retraite.xlsx",skiprows=3)
df= df.iloc[0:18,:]
```

```python
df =df.rename(columns={"Rapport démographique2":"rapport-retraite"}).drop('Retraités', axis=1).drop('Cotisants1', axis=1)
```

```python
df
```

```python
nump=df.to_numpy()
X = nump[:,:1]
y = nump[:,1]
X
```

```python
poly_deg = PolynomialFeatures(degree=3)
X_poly = poly_deg.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size = .3, random_state = 2)
```

```python
reg3 = LinearRegression()
reg3.fit(X_train, y_train)
```

```python
print(reg3.score(X_test, y_test))
print(reg3.score(X_train, y_train))
```

```python
years =[1994,1995,1997,1999,2002,2022,2024]
predictions = reg3.predict(poly_deg.fit_transform([[1994], [1995], [1997], [1999],[2002],[2022],[2024]]))
predictions
for i in  range(len(predictions)):
    df.loc[len(df.index)] =[years[i],predictions[i]]
```

```python
df = df.sort_values("Année")
df
```

```python
departement_code= ['16','17','19','23','24','33','40','47','64','79','86','87']
df_passing_dep = []
for dep in departement_code :
    df_dep = df.copy()
    df_dep["Code du département"] = dep
    df_passing_dep.append(df_dep)
len(df_passing_dep)
```

```python
df_passing_dep[7]
```

```python
df_concat =  pd.concat(df_passing_dep)
df_concat
```
