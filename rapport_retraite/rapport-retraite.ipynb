```python
import pandas as pd
```

```python
df = pd.read_excel("rapport-retraite.xlsx",skiprows=3)
df= df.iloc[0:18,:]
```

```python
df =df.rename(columns={"Rapport démographique2":"rapport-retraite"}).drop('Retraités', axis=1).drop('Cotisants1', axis=1)
```

```python
departement_code= ['16','17','19','23','24','33','40','47','64','79','86','87']
df_passing_dep = []
for dep in departement_code :
    df_dep = df.copy()
    df_dep["Code du département"] = dep
    df_passing_dep.append(df_dep)
len(df_passing_dep)
```

```python
df_passing_dep[7]
```

```python
df_concat =  pd.concat(df_passing_dep)
df_concat
```

```python
df_concat.to_csv("rapport-retraite.csv")
```
