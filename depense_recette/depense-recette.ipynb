```python
import pandas as pd
```

```python
departement_code= ['16','17','19','23','24','33','40','47','64','79','86','87']
```

```python
df = pd.read_excel("depense-recette.xlsx")
df=df.transpose()
df.columns = df.iloc[0]
df = df.drop(df.index[0])
df["Année"] = df.index
df=df.reset_index(drop=True)
df= df.append(df[df["Année"].eq(2023)]).sort_values("Année").reset_index(drop=True)
df.iloc[46,3] =2024
#df.reindex(df.index.append(df.index[df["Année"]=="2023"])).sort_index()
df
```

```python
df_passing_dep = []
for dep in departement_code :
    df_dep = df.copy()
    df_dep["Code du département"] = dep
    df_passing_dep.append(df_dep)
len(df_passing_dep)
```

```python
df_concat =  pd.concat(df_passing_dep)
df_concat
```

```python
df_concat.to_csv("depense-recette.csv")
```
