```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
departement_code= ['16','17','19','23','24','33','40','47','64','79','86','87']
df_passing_dep = []
for dep in departement_code :
    file= "./nombre-emplois/"+dep+".xlsx"
    df= pd.read_excel(file,skiprows=3)
    df =df.rename(columns={"Unnamed: 1":"nbr_emplois","Période":"Année"})
    nump=df.to_numpy()
    X = nump[:,:1]
    y = nump[:,1]
    poly_deg = PolynomialFeatures(degree=2)
    X_poly = poly_deg.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size = .3, random_state = 2)
    reg3 = LinearRegression()
    reg3.fit(X_train, y_train)
    df.loc[len(df.index)] =[2024,reg3.predict(poly_deg.fit_transform([[2024]]))[0]]
    df["Code du département"] =dep
    df_passing_dep.append(df.sort_values("Année"))
len(df_passing_dep)
df_passing_dep[0]
```

```python
df_concat =  pd.concat(df_passing_dep)
df_concat
```
