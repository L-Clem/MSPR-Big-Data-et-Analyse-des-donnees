```python
import pandas as pd
import functools 
```

```python
dicT_dep ={
    "Charente-Maritime": "17",
    "Charente": "16",

"Corrèze" :"19",
"Creuse": "23",
"Dordogne": "24",
"Gironde" :"33",
"Landes": "40",
"Lot-et-Garonne": "47",
"Pyrénées-Atlantiques" :"64",
"Deux-Sèvres": "79",
    "Haute-Vienne": "87",
"Vienne": "86",

}
categories = {"- Ensemble -":"taux_chomage_ensemble","- Ensemble des 15 à 24 ans -":"taux_chomage_15_24","- Ensemble des 25 à 49 ans -":"taux_chomage_25_49","- Ensemble des 50 ans ou plus -":"taux_chomage_50_+","- Femmes -":"taux_chomage_femme","- Hommes -":"taux_chomage_homme"}
print('|'.join(dicT_dep))
```

```python
df = pd.read_excel("taux-chommage.xlsx")
```

```python
df
```

```python
df=df[df["Libellé"].str.contains('|'.join(dicT_dep))]
df.drop(["idBank","Dernière mise à jour","Période"],axis=1, inplace=True)
df
```

```python
def get_dep(row):
    for dep in dicT_dep:
        if dep  in  row:
            return dicT_dep[dep]
```

```python
df["Code du département"] = df["Libellé"].apply(get_dep)
df
```

```python
df[df["Libellé"].str.contains("Ensemble")]
```

```python
df_cat = []
for cat in categories:
    print(cat)
    df_ens = df[df["Libellé"].str.contains(cat)]
    
    df_ens=df_ens.drop(["Libellé"],axis=1)
    df_ens=df_ens.melt(['Code du département'], var_name='Année',value_name=categories[cat])
    print(len(df_ens)," pour la categorie ",cat)
    df_cat.append(df_ens)
    

```

```python
df_merged = functools.reduce(lambda  left,right: pd.merge(left,right,on=['Code du département',"Année"],
                                            how='outer'), df_cat)
df_merged
```

```python
df_merged.to_csv("taux_chomage.csv")
```
