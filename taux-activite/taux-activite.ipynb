```python
import pandas as pd
import functools 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
df_f = pd.read_excel("taux-activite.xlsx",sheet_name="Figure 2a",skiprows=3).rename(columns={"Unnamed: 0":"Ann√©e","15-24 ans":"taux-activite-15-24-femme","25-49 ans":"taux-activite-25-49-femme","50 √† 64 ans":"taux-activite-50-64-femme","65-69 ans":"taux-activite-65-69-femme"})
df_f = df_f.drop(["15-64 ans"],axis=1).iloc[0:48,:]
df_f
```

```python
df_h = pd.read_excel("taux-activite.xlsx",sheet_name="Figure 2b",skiprows=3).rename(columns={"Unnamed: 0":"Ann√©e","15-24 ans":"taux-activite-15-24-homme","25-49 ans":"taux-activite-25-49-homme","50 √† 64 ans":"taux-activite-50-64-homme","65-69 ans":"taux-activite-65-69-homme"})
df_h = df_h.drop(["15-64 ans"],axis=1).iloc[0:48,:]
df_h
```

```python
df_merged = functools.reduce(lambda  left,right: pd.merge(left,right,on=["Ann√©e"],
                                            how='outer'), [df_f,df_h])
df_merged
```

```python
departement_code= ['16','17','19','23','24','33','40','47','64','79','86','87']
df_passing_dep = []
for dep in departement_code :
    df= df_merged.copy()
    df["Code du d√©partement"] = dep
    df_passing_dep.append(df)
len(df_passing_dep)
df_passing_dep[3]
```

```python
df_concat =  pd.concat(df_passing_dep).reset_index(drop=True)
df_concat
```

```python
df_concat.to_csv("taux-activite.csv")
```
