```python
import pandas as pd
```

```python
df = pd.read_excel("fecondite.xlsx",skiprows=3).rename(columns={"Unnamed: 0":"Année","Indicateur conjoncturel de fécondité (1) (pour 100 femmes)":"indicateur-fecondite","Âge moyen des mères (2) (en années)":"age-moyen-mere"})
df= df.iloc[0:31,:]
df = df[["Année",'indicateur-fecondite', 'age-moyen-mere']]
df
```

```python
departement_code= ['16','17','19','23','24','33','40','47','64','79','86','87']
df_passing_dep = []
for dep in departement_code :
    df_dep = df.copy()
    df_dep["Code du département"] = dep
    df_passing_dep.append(df_dep)
len(df_passing_dep)
```

```python
df_concat =  pd.concat(df_passing_dep)
df_concat
```

```python
df_concat.to_csv("fecondite.csv")
```
