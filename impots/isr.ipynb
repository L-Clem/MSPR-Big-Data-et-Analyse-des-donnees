```python
def process(df: pd.core.frame.DataFrame, year: int):
    df.info()
    
    # Supression des colonnes inutiles
    
    df = df.drop('@odata.etag', axis=1)
    df = df.drop('ItemInternalId', axis=1)
    df = df.drop('Commune', axis = 1)
    
    # Ajout de l'ann√©e
    df['Annee'] = year
    
    # Suppression des valeurs n.c.
    
    for col in df.columns:
        df.drop(df[df[col]  == 'n.c.'].index, inplace = True)
        df.drop(df[df[col]  == '. '].index, inplace = True)
        df.drop(df[df[col]  == '#VALUE!'].index, inplace = True)
        

    
    # Convertion des types de colonne
    
    for col in ['Libelle_de_la_commune', 'Revenu_fiscal_de_reference_par_tranche']:
        df[col] = df[col].astype(str)
    
    for col in ['Nombre_de_foyers_fiscaux', 'Nombre_de_foyers_fiscaux_imposes', 'Traitements_et_salaires_Nombre_de_foyers_concernes', 'Retraites_et_pensions_Nombre_de_foyers_concernes']:
        df[col] = df[col].astype(int)
    
    for col in ['Revenu_fiscal_de_reference_des_foyers_fiscaux', 'Impot_net_total', 'Revenu_fiscal_de_reference_des_foyers_fiscaux_imposes', 'Traitements_et_salaires_Montant', 'Retraites_et_pensions_Montant']:
        df[col] = df[col].astype(float)

    df.to_csv('df_sir_com_' + str(year) + '.csv')
```

```python
df_sir_com_2001 = process(df_sir_com_2001, 2001)
df_sir_com_2002 = process(df_sir_com_2002, 2002)
df_sir_com_2003 = process(df_sir_com_2003, 2003)
df_sir_com_2004 = process(df_sir_com_2004, 2004)
df_sir_com_2005 = process(df_sir_com_2005, 2005)
df_sir_com_2006 = process(df_sir_com_2006, 2006)
df_sir_com_2008 = process(df_sir_com_2008, 2008)
df_sir_com_2009 = process(df_sir_com_2009, 2009)
df_sir_com_2011 = process(df_sir_com_2011, 2011)
df_sir_com_2012 = process(df_sir_com_2012, 2012)
df_sir_com_2013 = process(df_sir_com_2013, 2013)
df_sir_com_2014 = process(df_sir_com_2014, 2014)
df_sir_com_2015 = process(df_sir_com_2015, 2015)
df_sir_com_2016 = process(df_sir_com_2016, 2016)
df_sir_com_2017 = process(df_sir_com_2017, 2017)
df_sir_com_2019 = process(df_sir_com_2019, 2019)
df_sir_com_2020 = process(df_sir_com_2020, 2020)
```
