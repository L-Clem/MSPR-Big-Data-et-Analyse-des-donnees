```python
import pandas as pd
```

```python
years_available = [2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
departement_code= [16,17,19,23,24,33,40,47,64,59,86,87]
```

```python
df_quartier = pd.read_excel('insecurite_quartier.xlsx')
df_renonce = pd.read_excel('insecurite_renonce.xlsx')
df_domicile = pd.read_excel('insecurite_domicile.xlsx')
```

```python
serie_ensemble_quartier =df_quartier[df_quartier["Tranche d'age"]=="Ensemble"].T.iloc[:,0]
serie_ensemble_renonce =df_renonce[df_quartier["Tranche d'age"]=="Ensemble"].T.iloc[:,0]
serie_ensemble_domicile =df_domicile[df_quartier["Tranche d'age"]=="Ensemble"].T.iloc[:,0]
```

```python
def create_dataframe_insecurite(serie_ensemble,type_insecurite):
    insecurite_femme = "Insecurite Femme " + type_insecurite
    insecurite_homme = "Insecurite Homme " + type_insecurite
    df = pd.DataFrame(columns=['Année','Code du département',insecurite_femme,insecurite_homme])
    for i in range(1,len(serie_ensemble),2):
        for y in range(len(years_available)):
            for z in range(len(departement_code)):
                row = {"Année":years_available[y],"Code du département":departement_code[z],insecurite_femme:serie_ensemble[i],insecurite_homme:serie_ensemble[i+1]}
                df.loc[len(df)] = row
    return df.reset_index()
```

```python
df_clean_quartier =create_dataframe_insecurite(serie_ensemble_quartier,"quartier")
df_clean_renonce =create_dataframe_insecurite(serie_ensemble_renonce,"renonce")
df_clean_domicile =create_dataframe_insecurite(serie_ensemble_domicile,"domicile")
```

```python
print(df_clean_renonce)
```

```python
df_insecurite_merged = pd.merge(df_clean_quartier, df_clean_renonce)
df_insecurite_merged = pd.merge(df_insecurite_merged,df_clean_domicile)
df_insecurite_merged
```

```python
df_insecurite_merged.to_csv("insecurite.csv")
```
