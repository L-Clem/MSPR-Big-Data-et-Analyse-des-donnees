```python
import pandas as pd
```

```python
departement_code= ['16','17','19','23','24','33','40','47','64','79','86','87']
df_passing_dep = []
for dep in departement_code :
    file_h= "Esperance_vie_naissance/H-"+dep+".xlsx"
    file_f = "Esperance_vie_naissance/F-"+dep+".xlsx"
    df_h = pd.read_excel(file_h,skiprows=3)
    df_f = pd.read_excel(file_f,skiprows=3)
    
    df = pd.concat([df_h.rename(columns={"Unnamed: 1":"esperance_vie_homme"}),df_f.drop('Période', axis=1)],axis=1).rename(columns={"Unnamed: 1":"esperance_vie_femme"})
    df =df.rename(columns={"Période":"Année"})
    df=df.append(df[df["Année"].eq(2023)]).sort_values("Année").reset_index(drop=True)
    df.iloc[49,0] = 2024
    df["Code du département"] =dep
    df_passing_dep.append(df)
len(df_passing_dep)
```

```python
df_concat =  pd.concat(df_passing_dep)
df_concat
```

```python
df_concat.to_csv("esperance_vie.csv")
```
