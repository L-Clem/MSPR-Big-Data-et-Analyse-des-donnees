```python
import pandas as pd
import functools 
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
years_election =[1994, 1995, 1997, 1999, 2002, 2004, 2007, 2009, 2012, 2014, 2017,
       2019, 2022,2024]
```

```python

df_sex = pd.read_excel("inegalite-sexe.xlsx",skiprows=3)
df_sex=df_sex.rename(columns={"Unnamed: 0":"Année","Unnamed: 3":"raport-inegalite-homme-femme"}).drop(["Femmes","Hommes"],axis=1)
df_sex= df_sex.iloc[0:19,:]
display(df_sex)

```

```python
df_soc = pd.read_excel("inegalite-sociale.xlsx",skiprows=3)

df_soc=df_soc.rename(columns={"Unnamed: 0":"Année","Unnamed: 6":"raport-inegalite-cadre-ouvrier"}).drop(["Ouvriers peu qualifiés ","Ouvriers qualifiés","Employés ","Professions intermédiaires ","Cadres "],axis=1)
df_soc= df_soc.iloc[0:19,:]
display(df_soc)
```

```python
df_rap = pd.read_excel("inegalite-rapport-interdecile.xlsx",skiprows=2)
df_rap= df_rap.iloc[0:3,:]
df_rap =df_rap.set_index('Rapport interdécile').transpose()
df_rap["Année"]= df_rap.index
df_rap=df_rap.rename(columns={"Rapport interdécile (D9/D1) ":"rapport-inegalite-d9-d1","Rapport interdécile (D9/D5) ":"rapport-inegalite-d9-d5","Rapport interdécile (D5/D1)":"rapport-inegalite-d5-d1"})
df_rap=df_rap.reset_index(drop=True)
display(df_rap)
df_rap.columns
```

```python
df = functools.reduce(lambda  left,right: pd.merge(left,right,on=["Année"],
                                            how='outer'), [df_sex,df_soc,df_rap])
df
```

```python
nump=df.to_numpy()
X = nump[:,:1]
y = nump[:,1:]
y
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_deg = PolynomialFeatures(degree=2)
X_poly = poly_deg.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size = .3, random_state = 2)
```

```python
reg3 = LinearRegression()
reg3.fit(X_train, y_train)
```

```python
print(reg3.score(X_test, y_test))
print(reg3.score(X_train, y_train))
```

```python
reg3.predict(poly_deg.fit_transform([[1994], [1995], [1997], [1999],[2022],[2024]]))
```

```python
years =[1994,1995,1997,1999,2022,2024]
predictions = reg3.predict(poly_deg.fit_transform([[1994], [1995], [1997], [1999],[2022],[2024]]))
for i in  range(len(predictions)):
    df.loc[len(df.index)] =np.insert(predictions[i], 0,years[i], axis=0)
    #predictions[i].insert(0, years[i])  
```

```python
df = df.sort_values("Année")
df
```

```python
departement_code= ['16','17','19','23','24','33','40','47','64','79','86','87']
df_passing_dep = []
for dep in departement_code :
    df_dep = df.copy()
    df_dep["Code du département"] = dep
    df_passing_dep.append(df_dep)
len(df_passing_dep)
```

```python
df_concat =  pd.concat(df_passing_dep).reset_index(drop=True)
df_concat
```

```python
df_concat.to_csv("inegalite.csv")
```
