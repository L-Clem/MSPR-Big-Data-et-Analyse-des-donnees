```python
import pandas as pd
import functools 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
dicT_dep ={
    "Charente-Maritime": "17",
    "Charente": "16",

"Corrèze" :"19",
"Creuse": "23",
"Dordogne": "24",
"Gironde" :"33",
"Landes": "40",
"Lot-et-Garonne": "47",
"Pyrénées-Atlantiques" :"64",
"Deux-Sèvres": "79",
    "Haute-Vienne": "87",
"Vienne": "86",

}
categories = {"Densité des médecins généralistes - Pour 100.000 habitants -":"densite_medecin_general","Densité des infirmiers diplômés d'État - Pour 100.000 habitants -":"densite_infirmier","Densité des médecins spécialistes - Pour 100.000 habitants -":"densite_sepcialiste"}
```

```python
df = pd.read_excel("densite-medecin.xlsx")
df
```

```python
df=df[df["Libellé"].str.contains('|'.join(dicT_dep))]
df.drop(["idBank","Dernière mise à jour","Période"],axis=1, inplace=True)
df
```

```python
def get_dep(row):
    for dep in dicT_dep:
        if dep  in  row:
            return dicT_dep[dep]
```

```python
df["Code du département"] = df["Libellé"].apply(get_dep)
df
```

```python
df_cat = []
for cat in categories:
    print(cat)
    df_ens = df[df["Libellé"].str.contains(cat)&df["Libellé"].str.contains("Série arrêtée")]
    print(len(df_ens),"vant")
    df_ens=df_ens.drop(["Libellé"],axis=1)
    df_ens=df_ens.melt(['Code du département'], var_name='Année',value_name=categories[cat])
    print(len(df_ens)," pour la categorie ",cat)
    df_cat.append(df_ens)
display(df_cat[1][(df_cat[1]["Code du département"]=="86")])
```

```python
df_merged = functools.reduce(lambda  left,right: pd.merge(left,right,on=['Code du département',"Année"],
                                            how='outer'), df_cat)
df_merged
```

```python
df_merged.to_csv("densite_medecin.csv")
```
