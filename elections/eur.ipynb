```python
df_eur_2019 = pd.read_excel('eur_2019.xls')
df_eur_2019 = df_eur_2019[df_eur_2019["Code du département"].isin(departement_code)]
print(len(df_eur_2019.columns.tolist()))
print(df_eur_2019.columns.tolist())

```

```python
df_eur_2019['Unnamed: 253']
```

```python
def test(row,begining,end,step,parti_column_name,step_column,voix_column_name,step_voix):
    political_parties = {
    "Extreme Gauche" : 0,
    "Gauche" : 0,
    "Droite" :0,
    "Extreme Droite" : 0,
    "Autre" : 0,
    }
    
    partie= DICT_POLITIC.get(row["Libellé Abrégé Liste"])
    if (partie):
        political_parties[partie] =  political_parties.get(partie) + float(row["% Voix/Exp"])
    else :
        print(row["Libellé Abrégé Liste"],"cae partie na pas de correspondance")
    for i in np.arange(begining,end,step):
        partie= DICT_POLITIC.get(row[parti_column_name+str((i+step_column))])
        if (partie):
            political_parties[partie] =  political_parties.get(partie) + float(row[voix_column_name+str((i+step_voix))])
        else :
            print(row[parti_column_name+str((i+step_column))],"bizzare ","cae partie na pas de correspondance")
    
    return political_parties.get("Extreme Gauche"),political_parties.get("Gauche"),political_parties.get("Droite"),political_parties.get("Extreme Droite"),political_parties.get("Autre")
       
        
        
    
    
```

```python
df_eur_2019[["Extreme Gauche","Gauche","Droite","Extreme Droite","Autre"]] = df_eur_2019.apply(test,begining=23,end=253,parti_column_name="Unnamed: ",voix_column_name="Unnamed: ",step_column=1,step_voix=6,step=7,axis=1, result_type="expand")
df_eur_2019[["Extreme Gauche","Gauche","Droite","Extreme Droite","Autre"]]
df_eur_2019 = df_eur_2019[['Code du département','Inscrits', 'Abstentions','Votants',"Extreme Gauche","Gauche","Droite","Extreme Droite","Autre"]]
df_eur_2019["Election"] = "Europeenne"
df_eur_2019["Année"] = "2019"
df_eur_2019
```
