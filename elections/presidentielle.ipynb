```python
df_pres_2022 = pd.read_excel('pres_2022.xlsx')
```

```python
print(df_pres_2022.columns.tolist())
```

```python
departement_code= ["16","17","19","23","24","33","40","47","64","59","86","87"]
departement_code_int= [16,17,19,23,24,33,40,47,64,59,86,87]
```

```python
def get_exp(exp):
    if isinstance(exp, str):
        return float(exp.replace(',',"."))
    return exp


def test(row,begining,end,step,parti_column_name,step_column,voix_column_name,step_voix):
    political_parties = {
    "Extreme Gauche" : 0,
    "Gauche" : 0,
    "Droite" :0,
    "Extreme Droite" : 0,
    "Autre" : 0,
    }
    
    partie= DICT_POLITIC.get(row["Nom"])
    
    if (partie):

        political_parties[partie] =  political_parties.get(partie) + get_exp(row["% Voix/Exp"])
    else :
        print(row["Nom"],"cae partie na pas de correspondance")
    for i in np.arange(begining,end,step):
        partie= DICT_POLITIC.get(row[parti_column_name+str((i+step_column))])
        if (partie):
            political_parties[partie] =  political_parties.get(partie) + get_exp(row[voix_column_name+str((i+step_voix))])
        else :
            print(row[parti_column_name+str((i+step_column))],"bizzare ","cae partie na pas de correspondance")
    
    return political_parties.get("Extreme Gauche"),political_parties.get("Gauche"),political_parties.get("Droite"),political_parties.get("Extreme Droite"),political_parties.get("Autre")
       
        
        
    
    
```

```python
df_pres_2022 = df_pres_2022[df_pres_2022["Code du département"].isin(departement_code)]
```

```python
df_pres_2022[["Extreme Gauche","Gauche","Droite","Extreme Droite","Autre"]]= df_pres_2022.apply(test,begining=23,end=89,parti_column_name="Unnamed: ",voix_column_name="Unnamed: ",step_column=1,step_voix=5,step=6,axis=1, result_type="expand")
df_pres_2022[["Extreme Gauche","Gauche","Droite","Extreme Droite","Autre"]]
df_pres_2022 = df_pres_2022[['Code du département','Libellé du département','Votants',"Extreme Gauche","Gauche","Droite","Extreme Droite","Autre"]]
df_pres_2022["Election"] = "Présidentielle"
df_pres_2022["Année"] = "2022"
df_pres_2022
```
